```python
def graph_foc(data1, data2):
    # Define plot space
    fig, ax = plt.subplots(figsize=(10, 5)) #ini untuk ukuran gambarnya
    x1 = total_foc(data1)[1]
    x2 = total_foc(data2)[1]
    y1 = total_foc(data1)[0]
    y2 = total_foc(data2)[0]
    # Define x and y axes
    ax.plot(x1, y1, color='blue', label='FOC dengan pertimbangan gelombang', linewidth=2, linestyle='-', marker='o')  # plot first line
    ax.plot(x2, y2, color='red', label='FOC tanpa pertimbangan gelombang', linewidth=2, linestyle='-.', marker='o')  # plot second line
    # Set plot title and axes labels
    ax.set_title("Fuel Oil Consumption", fontsize=15) 
    ax.set_xlabel("distance (nm)", fontsize=12)
    ax.set_ylabel("foc (ton)", fontsize=12) 
    plt.legend()
    plt.show()
    
def total_foc(route):
    list_foc = []
    list_distance = []
    x = 0
    y = 0
    for distance in route['Distance (nm)']:
        x += distance
        list_distance.append(x)
    for foc in route['foc (ton)']:
        y += foc
        list_foc.append(y)
    return list_foc, list_distance
```

```python
def graph_resistance(data1, data2):
    # Define plot space
    fig, ax = plt.subplots(figsize=(10, 5)) #ini untuk ukuran gambarnya
    x1 = total_rt(data1)[1]
    x2 = total_rcw(data2)[1]
    y1 = data1['R Total (kN)']
    y2 = data2['R Calm Water (kN)']
    # Define x and y axes
    ax.plot(x1, y1, color='cyan', label='resistance total', linewidth=2, linestyle='-', marker='o')  # plot first line
    ax.plot(x2, y2, color='red', label='resistance calm water', linewidth=2, linestyle='-.', marker='o')  # plot second line
    # Set plot title and axes labels
    ax.set_title("Resistance Total vs Resistance Calm Water", fontsize=15) 
    ax.set_xlabel("distance (nm)", fontsize=12)
    ax.set_ylabel("Resistance", fontsize=12) 
    plt.legend()
    plt.show()
    
def total_rt(route):
    list_rt = []
    list_distance = []
    x = 0
    y = 0
    for distance in route['Distance (nm)']:
        x += distance
        list_distance.append(x)
    for rt in route['R Total (kN)']:
        y += rt
        list_rt.append(y)
    return list_rt, list_distance

def total_rcw(route):
    list_rcw = []
    list_distance = []
    x = 0
    y = 0
    for distance in route['Distance (nm)']:
        x += distance
        list_distance.append(x)
    for rcw in route['R Calm Water (kN)']:
        y += rcw
        list_rcw.append(y)
    return list_rcw, list_distance
```
